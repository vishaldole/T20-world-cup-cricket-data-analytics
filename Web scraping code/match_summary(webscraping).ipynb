```python
#importing required libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager
import time
import re
import os
import pandas as pd
from bs4 import BeautifulSoup
```

```python
s = Service("C:\msedgedriver.exe")
driver = webdriver.Edge(service= s)
driver.get("https://stats.espncricinfo.com/ci/engine/records/team/match_results.html?id=2022%2F23;trophy=89;type=season")
#page url we need to scrape
driver.maximize_window()
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml') #parsing the webpage
```

```python
all_data = soup.find_all("tr",{"class":"data1"}) #class continer which contains all the required data
```

```python
len(all_data)
```

```python
Team1= []
Team2= []
Winner= []
Margin= []
Venue= []
Match_Date= []
Match_id= []
match_summary_df = pd.DataFrame()
```

```python
for data in all_data:
    n= data.find_all("td",{"nowrap":"nowrap"})
    x = [i.text for i in n]
    Team1.append(x[0])
    Team2.append(x[1])
    Winner.append(x[2])
    Margin.append(x[3])
    Match_Date.append(x[4])
    Match_id.append(x[5])
#collecting required data    
```

```python
for data in all_data:
    y = data.find_all("td",{"class":"left"})
    y_text = [i.text for i in y]
    venue= y_text[4]
    Venue.append(venue)
#collecting match venues    
```

```python
#Inserting values in dataframe
match_summary_df.insert(loc=0,column="Team1", value= Team1)
match_summary_df.insert(loc=1,column="Team2", value= Team2)
match_summary_df.insert(loc=2,column="Winner", value= Winner)
match_summary_df.insert(loc=3,column="Margin", value= Margin)
match_summary_df.insert(loc=4,column="Match_Date", value= Match_Date)
match_summary_df.insert(loc=5,column="Match_id", value= Match_id)
match_summary_df.insert(loc=4,column="Venue", value= Venue) 
```
